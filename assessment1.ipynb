```python
#new column addition
df["Age_Group"] = np.where(df["age"] >= 60,
                           "Senior",
                           "Adult")
```

```python
df.head()
```

```python
df["Heart_Function"] = np.where(df["ejection_fraction"] >= 50,
                                "Normal",
                                "Reduced")
```

```python
#Concatenation of DataFrames
df1 = df.iloc[:150]
df2 = df.iloc[150:]
```

```python
combined = pd.concat([df1, df2])
combined.head()
```

```python
patient_info = df[["age","sex"]]

medical_info = df[["diabetes","smoking"]]
```

```python
horizontal = pd.concat([patient_info, medical_info], axis=1)

horizontal.head()
```

```python
#Merging DataFrames
patient = df[["age","sex"]].copy()
medical = df[["diabetes",
              "smoking",
              "DEATH_EVENT"]].copy()
patient["Patient_ID"] = range(1, len(patient)+1)

medical["Patient_ID"] = range(1, len(medical)+1)
```

```python
merged = pd.merge(patient,
                  medical,
                  on="Patient_ID")
merged.head()
```
